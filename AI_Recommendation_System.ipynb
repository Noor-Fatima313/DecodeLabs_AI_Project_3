## Creating the Recommendation Dataset

A small dataset is created using a Python dictionary.

Each category contains items that belong to that interest.

```python
recommendations = {
    "technology": [
        "Python Programming",
        "Artificial Intelligence",
        "Machine Learning",
        "Data Science"
    ],

    "sports": [
        "Football",
        "Cricket",
        "Basketball",
        "Tennis"
    ],

    "movies": [
        "Inception",
        "Interstellar",
        "The Dark Knight",
        "Avengers: Endgame"
    ],

    "music": [
        "Pop Music",
        "Classical Music",
        "Rock Music",
        "Jazz Music"
    ],

    "books": [
        "Atomic Habits",
        "Deep Work",
        "The Alchemist",
        "Rich Dad Poor Dad"
    ]
}
```

## Displaying Available Interests

The user can choose one or more interests from the available categories.

```python
print("Available Interests:\n")

for category in recommendations:
    print("-", category.title())
```

## Taking User Input

The user enters one or more interests separated by commas.

Example:

technology,movies

```python
user_input = input("\nEnter your interests (comma separated): ")

user_interests = [
    interest.strip().lower()
    for interest in user_input.split(",")
]
```

## Recommendation Function

This function checks whether each user interest exists in the dataset.

If a match is found, all related items are recommended.

```python
def recommend_items(interests):

    recommended = []

    for interest in interests:

        if interest in recommendations:

            recommended.extend(recommendations[interest])

    return recommended
```

## Displaying Recommendations

The recommendation function is called and the matching items are displayed.

```python
result = recommend_items(user_interests)

print("\nRecommended Items\n")

if result:

    for item in result:
        print("•", item)

else:

    print("No matching interests found.")
```
